```python
import pandas as pd
df = pd.read_csv('diabetes.csv')
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('Outcome', axis='columns')
y = df['Outcome']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
from tensorflow import keras
```

```python
model = keras.models.Sequential()
```

```python
model.add(keras.layers.Dense(32, input_dim = 8, activation = 'relu'))
```

```python
model.add(keras.layers.Dense(8, activation = 'relu'))
```

```python
model.add(keras.layers.Dense(1, activation = 'sigmoid'))
```

```python
#flatten, relu, 16 neurons
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(x_train, y_train, epochs = 50, batch_size=32)
```

```python
model.predict(x_test.head(1))
```

```python
x_test.head(1)
```

```python
y_test.head(1)
```

```python
model.evaluate(x_test, y_test)
```

```python
from tensorflow.keras.datasets import mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
import matplotlib.pyplot as plt
plt.matshow(x_train[0])
```

```python
model = tensorflow.keras.models.Sequential()
model.add(tensorflow.keras.layers.Flatten(input_shape=(28,28)))
```

```python
model.add(tensorflow.keras.layers.Dense(50, activation='relu'))
model.add(tensorflow.keras.layers.Dense(10, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss = 'sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs =10)
```

```python
model.evaluate(x_test, y_test)
```
