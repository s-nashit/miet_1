```python
import pandas as pd
```

```python
df = pd.DataFrame({'marks' : [72, 43,49,65,81,12,39],
'status' : ['Pass', 'Pass', 'Pass', 'Pass', 'Pass', 'Fail', 'Fail']})
```

```python
df.head()
```

```python
df['status'] = df['status'].apply(lambda x: 1 if x == 'Pass' else 0)
```

```python
df.tail(1)
```

```python
x = df.drop('status', axis='columns')
y = df['status']
```

```python
from sklearn import svm
```

```python
model = svm.SVC()
```

```python
model.fit(x,y)
```

```python
model.predict([[12]])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(3,3))
plt.scatter(df['marks'], df['status'])
```
