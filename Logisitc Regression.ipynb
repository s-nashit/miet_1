```python
import pandas as pd
df = pd.read_csv('c.csv')
```

```python
df
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
x = df.drop('attendance', axis='columns')
y = df['distinction']
```

```python
model = LogisticRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[87]])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(3,3))
plt.scatter(df['attendance'], df['distinction'])
```

```python
import math
def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
print(round(sigmoid(1000)))
```

```python
m = model.coef_
c = model.intercept_
```

```python
def pred(att):
    y = m*att+c
    z = sigmoid(y)
    return z
    
```

```python
pred(87)
```

```python
plt.figure(figsize=(3,3))
plt.scatter(df['income'], df['loan'])
```

```python
model = LogisticRegression()
```

```python
x = df.drop('loan', axis='columns')
y = df['loan']
```

```python
model.fit(x,y)
```

```python
pred = model.predict([[61000]])
```

```python
if pred == 1:
    print('Loan approved')
else:
    print('No loan')
```

```python
#y = mx+c
m = model.coef_
c = model.intercept_
```

```python
def sigmoid(a):
    return 1/(1+math.exp(-a))
```

```python
round(sigmoid(-98))
```

```python
def pred(income):
    y = m*income+c
    z = sigmoid(y)
    return(z)
    
```

```python
round(pred(59000))
```
