```python
import pandas as pd
```

```python
df = pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
df.drop('College_ID', axis='columns', inplace=True)
```

```python
df.head(5)
```

```python
df['Internship_Experience']=df['Internship_Experience'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
df.head()
```

```python
df['Placement'] = df['Placement'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
df.head()
```

```python
x = df.drop('Placement', axis='columns')
y = df['Placement']
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x,y)
```

```python
x.head(1)
```

```python
iq = int(input('enter iq'))
prev = float(input('enter previous result'))
cgpa = float(input('enter cgpa'))
acad = int(input('enter academic performance'))
inter = int(input('enter internship numbers'))
eca = int(input('enter eca'))
cs = int(input('enter communication skills'))
projects = int(input('no of projects completed'))
pred = model.predict([[iq, prev, cgpa,acad,inter,eca,cs,projects]])
if pred ==1:
    print('Placed')
else:
    print('Not')
```
