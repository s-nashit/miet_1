```python
hours = [1, 2, 3, 4, 5, 6]
marks = [40, 43, 49, 53, 60, 70]
```

```python
#import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(2,2))
plt.scatter(hours, marks)
plt.xlabel('Hours')
plt.ylabel('Marks')
plt.title('Marks Obtained')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
import numpy as np
x = np.array(hours).reshape(-1,1)
y = np.array(marks)
```

```python
x
```

```python
y
```

```python
model.fit(x, y)
```

```python
model.predict([[3.5]])
```

```python
m = model.coef_
```

```python
c = model.intercept_
```

```python
#y =mx+c
```

```python
m*3.5 +c
```
